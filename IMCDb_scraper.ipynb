```python
# dependencies
import pandas as pd
import time
from datetime import date
from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# export dataframe of results with today's date
today = date.today()
d_today = today.strftime("%Y_%m_%d")
```

```python
top_100df = pd.read_csv("output_data/Top100_Movies_2010-2021.csv")
```

```python
# create list of years
years = top_100df["Year"].unique().tolist()
print(years)
```

```python
# use a dictionary to create multiple empty lists to store each  year's top 100 movies
obj = {}
for year in years:
    obj[year] = []
print(obj)
```

```python
# fill dictionary with lists of the top 100 movies per year, display a sample year
for year in obj:
    top_100 = top_100df.loc[top_100df["Year"] == year]
    obj[year] = top_100["Release"].tolist()
print(obj[2016])
```

```python
# setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
def car_mov(titles):
    movie_dict = {}
    for title in titles:
        movie_dict[title] = []
    for title in titles:
        url = "https://www.imcdb.org/"
        browser.visit(url)
    
        browser.find_by_css('input').fill(title)
        time.sleep(1)
        browser.find_by_value('Search').click()
        time.sleep(5)
    
        html = browser.html
        soup = bs(html, 'html.parser')
    
        results = soup.find_all('div', class_='ThumbnailBox')
        if len(results) != 0:
            print("Everything good.")
            good = 1
        else:
            good = 0
            print("Not good yet...")
            try:
                browser.links.find_by_text(title).click()
                time.sleep(5)
            except:
                print("No data for the movie.")
                good = 0
                continue
        
        if good == 1:
            good2 = 1
            print("Everything still good.")
        else:
            html = browser.html
            soup = bs(html, 'html.parser')
            results = soup.find_all('div', class_='ThumbnailBox')
            if len(results) != 0:
                good2 = 1
                print("Everything good now.")
            else:
                good2 = 0
                print("Not good at all: no car data.")
                continue
        
        if good2 == 1:
            print("Parsing through results...")
            count = 0
            for result in results:
                try:
                    x = result.find('span', class_='Stars')
                    y = len(x)
                    z = result.find('a')['href']
                    vehicles = result.find_all('a')
                    vehicles.pop(0)
                    car_info = []
                    for ele in vehicles:
                        ele_text = ele.text
                        car_info.append(ele_text)
                    car_infos = " "
                    info_string = car_infos.join(car_info)
                    count += 1
                    print(f"{count}: {info_string}, https://www.imcdb.org/{z}, {y} Star(s)")
                    movie_dict[title].append([count, info_string, z, y])
                except:
                    y = 'Nan'
                    z = result.find('a')['href']
                    vehicles = result.find_all('a')
                    vehicles.pop(0)
                    car_info = []
                    for ele in vehicles:
                        ele_text = ele.text
                        car_info.append(ele_text)
                    car_infos = " "
                    info_string = car_infos.join(car_info)
                    count += 1
                    print(f"{count}: {info_string}, https://www.imcdb.org/{z}, {y} Star(s)")
                    movie_dict[title].append([count, info_string, z, y])
        else:
            pass
    browser.quit()
    return movie_dict
```

```python
top_2016 = obj[2016]
top_2016cars = car_mov(top_2016)
```

```python
# create dataframe of data for each vehicle, with movie and movie year included
title = []
year = []
auto = []
link = []
stars = []
for movie in top_2016:
    try:
        cars = top_2016cars[movie]
        for car in cars:
            title.append(movie)
            year.append(2016)
            auto.append(car[1])
            link.append(car[2])
            stars.append(car[3])
    except:
        print("No cars for this movie found.")
```

```python
top_2016_df = pd.DataFrame({"Release": title, "Year": year, "Car": auto, "url": link, "Stars": stars})
```
